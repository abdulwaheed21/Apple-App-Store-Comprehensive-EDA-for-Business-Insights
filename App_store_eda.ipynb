### **Introduction**

This notebook presents an Exploratory Data Analysis (EDA) of a dataset containing over 1.23 million app entries from the Apple App Store. The data includes various features of mobile applications such as app names, genres, content ratings, sizes, versions, prices, developer details, and user ratings. The goal of this EDA is to gain insights into the distribution of app attributes, identify trends, and uncover relationships between different features, which may provide valuable information about the app ecosystem on the Apple App Store. 

### **Dataset Description**

The dataset contains 21 columns, including: 
- **App_Id**: Unique identifier for the app.
- **App_Name**: Name of the app. 
- **AppStore_Url**: URL to the app's page on the App Store. 
- **Primary_Genre**: Genre/category of the app (e.g., Education, Games, etc.). 
- **Content_Rating**: The age rating of the app (e.g., 4+, 12+). 
- **Size_Bytes**: Size of the app in bytes. 
- **Required_IOS_Version**: Minimum iOS version required to run the app. 
- **Released**: The date the app was initially released. 
- **Updated**: The date the app was last updated. 
- **Version**: The current version of the app. 
- **Price**: Price of the app in USD. 
- **Currency**: Currency used for the app's price. 
- **Free**: Whether the app is free or not. 
- **DeveloperId**: Unique identifier for the developer. 
- **Developer**: Name of the developer.
- **Developer_Url**: URL to the developerâ€™s profile on the App Store. 
- **Developer_Website**: Developer's official website. 
- **Average_User_Rating**: Average user rating for the app. 
- **Reviews**: Number of reviews for the app. 
- **Current_Version_Score**: User rating for the current version of the app. 
- **Current_Version_Reviews**: Number of reviews for the current version. 

### **Objectives of the Analysis**

The objective of this EDA is to explore the key characteristics of apps available on the Apple App Store. We aim to derive insights that can help guide business decisions, such as which genres are most competitive, how pricing correlates with app size and ratings, and which genres attract the highest user engagement. By understanding the trends in app distribution, ratings, and pricing, we can identify areas of opportunity for launching new apps and improving existing offerings.

###  3. Handling Missing Data

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# Check the percentage of missing values in each column, in descending order
missing_data = df.isnull().sum().sort_values(ascending=False)
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_data_percentage = pd.DataFrame({'Missing Values': missing_data, 'Percentage': missing_percentage})

# Display columns with missing data and percentage
missing_data_percentage[missing_data_percentage['Missing Values'] > 0]
```

```python
# Visualize missing values using a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Drop columns with more than 25% missing data
threshold = 25  # Define the threshold percentage
columns_to_drop = missing_data_percentage[missing_data_percentage['Percentage'] > threshold].index

# Drop the columns
df.drop(columns=columns_to_drop, axis=1, inplace=True)

# Verify that the columns are dropped
df.info()
```

```python
# Fill missing values in numerical columns with the median value
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

for col in numerical_columns:
    df[col].fillna(df[col].mean(), inplace=True)

# Verify that numerical missing values are filled
df.isnull().sum()
```

```python
# Remove rows with missing values in remaining columns (non-numerical ones)
df.dropna(inplace=True)

# Verify that there are no missing values left
df.isnull().sum()
```

We have cleaned the dataset from null values

#### 5.1 Transforming Content Ratings into Nominal Categories

The **Content_Rating** column in our dataset represents age-based content categories such as "4+", "9+", "12+", and "17+". To make this more interpretable, we will transform these numerical age ratings into nominal categories that describe the target audience in more familiar terms. The following transformations will be applied:

- **4+**: Child (suitable for all ages)
- **9+**: Pre-Teen (suitable for ages 9 and above)
- **12+**: Teen (suitable for ages 12 and above)
- **17+**: Adult (suitable for ages 17 and above)
- **Not yet rated**: Unrated (not yet classified)

After the transformation, we will create a new column, `Content_Rating_Nominal`, containing these new descriptive categories. This transformation will help us better analyze and visualize app data based on content suitability for different age groups.


```python
# Create a dictionary to map the content ratings to nominal categories
content_rating_mapping = {
    '4+': 'Child',
    '9+': 'Pre-Teen',
    '12+': 'Teen',
    '17+': 'Adult',
    'Not yet rated': 'Unrated'
}

# Apply the mapping and replace the original 'Content_Rating' column
df['Content_Rating'] = df['Content_Rating'].map(content_rating_mapping)

# Check the transformation
df['Content_Rating'].head()

# Check the counts of the updated Content_Rating column
df['Content_Rating'].value_counts()
```
